```python
def people_cell(cell):
    if cell == 'n.a.':
        return 'harshit prajapati'
    return cell
```

```python
def eps_cell(cell):
    if cell == 'not available':
        return None
    return cell
```

```python
def price_cell(cell):
    if cell == 'n.a.':
        return None
    return cell
```

```python
df1 = pd.read_excel('Desktop/py-master/pandas/4_read_write_to_excel/stock_data.xlsx',converters = {
    'people': people_cell,
    'eps': eps_cell,
    'price':price_cell
})
df1
```
