```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Desktop/py-master/pandas/6_handling_missing_data_replace/weather_data.csv")
df
```

```python
df1 = df.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},np.NaN)
df1
```

```python
df1 = df.replace('[A-Za-z]','',regex=True) #used to chopoff extra data like 32F to 32 in temperature column
df1
```

```python
df1 = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed': -99999
},'',regex=True)
```
