```python
import pandas as pd
df = pd.read_csv("Desktop/py-master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv",
                parse_dates=['day'])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
```

```python
newdf = df.fillna(0)
newdf
```

```python
newdf = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event':'no event'
})
newdf
```

```python
newdf = df.fillna(method='ffill',limit =1) #fill na with previous values with limit as 1
newdf
```

```python
newdf1= df.fillna(method='bfill',axis='columns') #fill na with next values (here row by row)
newdf1
```

```python
newdf = df.fillna(0)
newdf = df.interpolate() #Using interpolate function to fillna with better guess with linear method
newdf
```

```python
newdf = df.fillna(0)
newdf = df.interpolate(method='time')
newdf
```

```python
dropdf = df.dropna() # drop rows with atleast one na values
```

```python
dropdf1 = df.dropna(how="all") # drop rows having all vlaues as na
dropdf1
```

```python
dropdf2 = df.dropna(thresh = 1)
dropdf2
```
